**Contexto**

La hepatitis, esencialmente, es una enfermedad que causa inflamación del hígado. Dado que el hígado es vital para combatir las infecciones, la hepatitis es muy perjudicial y provoca síntomas como orina oscura, fiebre, dolor y, con bastante frecuencia, la muerte. Hay muchos tipos de Hepatitis, pero utilizaré un conjunto de datos de pacientes con Hepatitis que examina el virus en general, sin profundizar en cada tipo específico. La hepatitis suele producirse por transmisión sexual o a través de alimentos y líquidos, y causa alrededor de 1,34 millones de muertes al año en todo el mundo.

## **Segunda Parte:** Preparación de los Datos

```python
colums_mode = ['Esteroides', 'Fatiga', 'Malestar',	'Anorexia',	'Higado_Graso',
               'Higado_Firme', 'Bazo_Palpable',	'Telangiectasias',	'Ascitis',
               'Varices',	'Bilirrubina', 'Fosfatasa_Alcalina',	'SGOT',
               'Albumina',	'Tiempo_Protrombina'
               ]
```

```python
# Remplazar los valores nulos por la moda
df.fillna(df[colums_mode].mode().iloc[0], inplace=True)
```

```python
df.info()
```

```python
# df.boxplot(column='Edad', vert=False)
# df.boxplot(column='Fosfatasa_Alcalina', vert=False)
# df.boxplot(column='SGOT', vert=False)

# df.boxplot(column='Tiempo_Protrombina', vert=False)
# df.boxplot(column='Bilirrubina', vert=False)
df.boxplot(column='Albumina', vert=False)
```

```python
df.loc[df['Edad'] > 71, 'Edad'] = round(df['Edad'].mean())
df.loc[df['Fosfatasa_Alcalina'] > 170, 'Fosfatasa_Alcalina'] = round(df['Fosfatasa_Alcalina'].mean())
df.loc[df['SGOT'] > 200, 'SGOT'] = round(df['SGOT'].mean())

df.loc[df['Tiempo_Protrombina'] > 100, 'Tiempo_Protrombina'] = round(df['Tiempo_Protrombina'].mean())
df.loc[df['Bilirrubina'] > 2.5, 'Bilirrubina'] = df['Bilirrubina'].mean()
df.loc[df['Albumina'] < 2.7, 'Albumina'] = df['Albumina'].mean()
df.loc[df['Albumina'] > 5, 'Albumina'] = df['Albumina'].mean()
```

```python
df
```

```python
# Transformar columnas categoricas en numericas
# Female: 0, Male: 1
# False:0, True: 1
```

```python
df['Sexo'] = df['Sexo'].map({'female': 0, 'male': 1})
```

```python
columns_transform = ['Esteroides', 'Antivirales',	'Fatiga',	'Malestar',	'Anorexia',
                     'Higado_Graso',	'Higado_Firme',	'Bazo_Palpable',	'Telangiectasias',
                     'Ascitis',	'Varices', 'Histología']
```

```python
for column in columns_transform:
    df[column] = df[column].map({False: 0, True: 1})
```

```python
df.to_excel('hepatitis_final_clean.xlsx', index=False)
```
